```python
def read_dataframe():
    
    train = pd.read_csv("train.csv")
    test = pd.read_csv("test.csv")
    
    nrow_train = train.shape[0]
    combined = pd.concat([train, test], sort= True)
    
    combined.reset_index(drop=True, inplace= True)
    
    return train, test, combined, nrow_train

train, test, combined, nrow_train = read_dataframe()
```

```python
def get_dummies(df, column_name):
    dummies = pd.get_dummies(df[column_name])
    
    return pd.concat([df, dummies], axis = 1)

combined = get_dummies(combined, 'residence_area_type')
combined = get_dummies(combined, 'sourcing_channel')

del combined['residence_area_type']
del combined['sourcing_channel']
del combined['id']

combined.head(7)
```

## Model Fitting and Prediction

Let's split the combined dataset back into train and test data.

```python
def split(combined):
    target_label = 'renewal'
    label = [lab for lab in combined.columns.tolist() if lab != target_label]
    
    train_idx = combined[combined.renewal.notnull()].index.tolist()
    test_idx = combined[combined.renewal.isnull()].index.tolist()
    
    X_train = combined.loc[train_idx, label].values
    X_test = combined.loc[test_idx, label].values
    y_train = combined[target_label].loc[train_idx].values
    
    return X_train, X_test, y_train

X_train, X_test, y_train = split(combined)
```

```python
from xgboost.sklearn import XGBClassifier 

xgb2 = XGBClassifier(
    objective = 'binary:logistic',
    min_child_weight = 0.5,
    max_depth = 6,
    colsample_bytree = 0.5,
    subsample = 1,
    gamma = 0.0,
    eta = 0.02,
    max_delta_step = 1.2,
    seed = 28)

xgb2.fit(X_train, y_train)  
xgb2_pred_prob = xgb2.predict_proba(X_test)
```

```python
renewal_prob = pd.DataFrame(xgb2_pred_prob[: ,1])
```

```python
from sympy import log

def get_hours(y):
    x = - 5*(log(1- (y/20)))
    return x

renewal_prob.apply(get_hours,axis=1)
```

```python
def get_incentive(y):
    x = - 400*(log(1- (y/100)))
    return x

renewal_prob = renewal_prob.apply(get_incentive,axis=1)
```

```python
df_id = pd.DataFrame(test['id'])
df_renewal = pd.DataFrame(xgb2_pred_prob[: ,1])
df_incentive= pd.DataFrame( renewal_prob )

submission= pd.concat([df_id,df_renewal,df_incentive], axis=1)
submission.columns = ['id', 'renewal', 'incentives']

submission.to_csv('submission.csv', index=False)
```
